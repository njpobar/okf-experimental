# E-commerce SQLite experimentation notebook

Use this notebook to inspect the seeded database and pull up to 100 rows from each table quickly.

Tables included:

- `categories`
- `customers`
- `products`
- `orders`
- `order_items`
- `reviews`


```python
from pathlib import Path
import sqlite3

import pandas as pd
from IPython.display import display

DB_PATH = Path("ecommerce.sqlite")
conn = sqlite3.connect(DB_PATH)

pd.set_option("display.max_rows", 120)
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 160)

def pull_rows(table_name: str, limit: int = 100, order_by: str | None = "id") -> pd.DataFrame:
    """Return up to `limit` rows from a table."""
    query = f'SELECT * FROM {table_name}'
    if order_by:
        query += f' ORDER BY {order_by}'
    query += f' LIMIT {int(limit)}'
    return pd.read_sql_query(query, conn)

def show_rows(table_name: str, limit: int = 100, order_by: str | None = "id") -> pd.DataFrame:
    df = pull_rows(table_name, limit=limit, order_by=order_by)
    display(df)
    return df
```

## Quick schema check

This confirms the tables that exist in the database before you start experimenting.

```python
schema_tables = pd.read_sql_query(
    """
    SELECT name AS table_name
    FROM sqlite_master
    WHERE type = 'table'
      AND name NOT LIKE 'sqlite_%'
    ORDER BY name
    """,
    conn,
)
display(schema_tables)
```

```python
out = pd.read_sql_query(
    """
    select *
    from categories
    limit 100
    """,
    conn
)
display(out)
```

## Handy join examples

A couple of starter queries for experiments once you want to move beyond raw table samples.

```python
orders_join = pd.read_sql_query(
    """
    SELECT
      o.id AS order_id,
      o.order_date,
      o.status,
      c.first_name || ' ' || c.last_name AS customer_name,
      p.name AS product_name,
      oi.quantity,
      oi.line_total_cents
    FROM orders o
    JOIN customers c ON c.id = o.customer_id
    JOIN order_items oi ON oi.order_id = o.id
    JOIN products p ON p.id = oi.product_id
    ORDER BY o.order_date DESC, o.id DESC
    LIMIT 100
    """,
    conn,
)
display(orders_join)
```
